```python
# write a python program to count
#no.of alaphabets,no.of words,no.of lines,
#no.of special chars and no.of digits
```

```python
freq=open("new.py",'r')
print(freq.read()) #str,readlines-li
```

```python
ac=wc=lc=spc=dc=0
freq=open("new.py",'r')
data=freq.readlines()
for li in data:
    lc+=1
    for w in li.split():
        wc+=1
        for ch in w:
            if ch.isalpha():
                ac+=1
            elif ch.isdigit():
                dc+=1
            else:
                spc+=1

print("no.of lines=",lc)
print("no.of words=",wc+1)
print("no.of alphabets=",ac)
print("no.of digits=",dc)
print("no.of special chars=",spc)
```
